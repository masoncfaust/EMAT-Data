I made this report to find out what accounts I like the most on instagram. I am trying to answer this question becasue I want to know what kinds of instagram pages I interact with the most and what kind of content I am the most interested in. This could end up explaining why I interact with instagram so much. My hypothisis is that I bet the most liked accounts on instagram are going to be animal related.

I got this data by downloading it from instagram.com in JSON form.

To start, I'm going import the tools I need, such as matplotlib, pyplot, datetime and seaborn.

```python
from matplotlib import pyplot as plt
import json
from datetime import datetime as dt
import seaborn
```

This is the command that loads the json data

```python
with open(r'C:\Users\mason\Desktop\mason_faust_357_20210126\likes.json') as l:
    likes = json.load(l)
```

This is the command that imports the tool pandas as well as turn the media likes section of the json file into a DataFrame object.

```python
import pandas as pd
likes_df = pd.DataFrame(likes['media_likes'])
```

This is a bit of code that then turns the DataFrame object into a organised table with a timestamp and account file.

```python
likes_df = pd.DataFrame(likes['media_likes'], columns = ['timestamp', 'account'])
likes_df [7:70]
```

This is a bit of code that organises the list to show how many times I liked a post from each account I've liked a post from.

```python
count_likes = likes_df.groupby('account').count()
count_likes[:10]
```

This bit of code gets the previous list and organises it even more to list each account in order from most to least amount of posts liked.

```python
count_likes.sort_values('timestamp', ascending = False)
```

The next few sections of code create graphs using the now organised data.

```python
sortcount = count_likes.sort_values('timestamp', ascending = False)
labels = sortcount['timestamp'].keys()
vals = sortcount['timestamp']
width = 6
```

```python
fig, ax = plt.subplots(figsize = (10, 10), dpi = 80, facecolor = 'w', edgecolor = 'k')
ax.bar(labels, vals, width)
ax.set_ylabel('Likes')
plt.xticks(rotation = 90, fontsize = 'small')
plt.show()
```

Becasue this bar graph is unreadable and I dont have the coding skills to fix it, I am choosing an alternitive approach by using a pie graph instead of a bar graph because I know how to make a pie graph provide a visual presentation of the data that people can easily see.

```python
toplikes = count_likes['timestamp'] > 500
countstop, countsbot = sortcount[toplikes], sortcount[-toplikes]
other = countsbot.sum()

piedat = countstop.append(pd.DataFrame(other['timestamp'], columns = ['timestamp'], index = ['Accounts less then 500 likes']))
piedat
```

```python
vals = piedat['timestamp']
labels = piedat['timestamp'].keys()


fig2, ax2 = plt.subplots()
ax2.pie(vals, labels = labels)
ax2.axis('equal')
plt.show()
```
